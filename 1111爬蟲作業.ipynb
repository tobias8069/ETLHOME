```python
import requests
import re
from bs4 import BeautifulSoup
import json

def find_job(keyword):
    
    url = 'https://www.1111.com.tw/job-bank/job-index.asp?si=1&ss=s&ks={}&ps=50'.format(keyword)
    
    page = 1
    
    lists = []   #每一筆工作內容
    count = 0    #計算工作數量
    
    while True:
        url_page = url + '&page=' + str(page)
        
        joblist = job_url_list(url_page)  #每一頁工作網址列表
        
        if joblist == []:    #只要工作列表沒有網址; break
            break
        
        for row in joblist:
            
            lists.append(job_content(row))
            
            #每一筆row的內容
            '''
            {'area': '新北市板橋區 雙十路二段10之1號8樓',
             'company': '默科資訊股份有限公司',
             'job': 'Java Developer',
             'skill': '',
             'update_date': '2018/1/20',
             'url': 'https://www.1111.com.tw/job/80057790/'}
            '''
            
            with open('1111JOB/{}page{}.json'.format(keyword,str(page)),'w') as f:  #存入json檔 用json.dump()
                      json.dump(lists,f)                                            #讀取json檔 用json.load()
                      #f.write(str(joblist)
              
        
        count += len(lists)    
        print('page' + str(page))
        print('總共輸入了' + str(count)+ '筆資料')       
              
        lists = []
        page += 1
             
                  
    
                        
def job_url_list(url):   #每一頁工作列表的url 
    
    '''
    傳回 list
    ['https://www.1111.com.tw/job/80197415/?agent=sticktop_71952995_80197415',
     'https://www.1111.com.tw/job/80057790/',
     'https://www.1111.com.tw/job/79828717/',
     'https://www.1111.com.tw/job/55576904/',
     'https://www.1111.com.tw/job/76821519/',]
    '''
    
    HOST = 'https://www.1111.com.tw/job/'
    
    resp = requests.get(url)
    joblist = re.findall('<a href="//www.1111.com.tw/job/(\d+/\S*)"',resp.text) #看正規表達式
    
    
    
    return [ HOST + s for s in joblist]

def job_content(url):
    '''
    {'area': '新北市板橋區 雙十路二段10之1號8樓',
     'company': '默科資訊股份有限公司',
     'job': 'Java Developer',
     'skill': '',
     'update_date': '2018/1/20',
     'url': 'https://www.1111.com.tw/job/80057790/'}
    
    '''
    
    resp =requests.get(url)
    #resp.content.deconde('utf-8) = resp.text

    soup = BeautifulSoup(resp.text,'lxml')
    content = {}
    content['url']  = url                                                                        #1111工作網址
    #content['job'] = soup.select('#menu > div > div.logoTitle > h1')[0].text                    #工作名稱？
    content['job']  = soup.select('#menu > div > div.logoTitle > h1')[0].text.split('面議')[0].split('月薪')[0].strip()
    content['company'] =soup.select('#menu > div > div.logoTitle > ul > li.ellipsis > a')[0].text #公司名稱
    content['area'] = soup.select('div.listContent')[0].text.replace('地圖','')                   #工作地點
    
    
    '''
    分清楚 nth-of-type 和 nth-of-child的差別
    content['skill'] = soup.select("""                                    
                                    article:nth-of-type(2) >             
                                    ul > li:nth-of-type(5) 
                                    > div.listContent
                                  """
                                  )[0].text.split('、')
    每個網頁的 li:nth-of-type(5) 不一定是電腦技能,或是沒有電腦技能
    '''
    #為要求條件的block
    a = soup.select("""                                    
                        article:nth-of-type(2) > 
                        ul > li
                    """)
    for row in a:
        if re.findall('電腦專長',row.text) != []:
            content['skill'] = row.text.split('：')[1].split('、')
        else: 
            content['skill'] = ''
            
    content['update_date'] = re.findall('更新日期：(\d+/\d+/\d+)',resp.text)[0]                                  
    
    return content
    
    
    
```
